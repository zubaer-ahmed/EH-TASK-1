```python
import json
import pandas as pd

with open('./data/districts_bd.json', 'r', encoding='utf-8') as file:
    json_data = json.load(file)
    districts = pd.DataFrame.from_dict(json_data)
with open('./data/postcodes_bd.json', 'r', encoding='utf-8') as file:
    json_data = json.load(file)
    postcodes = pd.DataFrame.from_dict(json_data)

postcodes.to_json('./data/output/postcodes_bd.json', orient='records', force_ascii=False)
postcodes
districts.to_json('./data/output/districts_bd.json', orient='records', force_ascii=False)
districts
```

```python
import json
import pandas as pd

services = []
with open('./data/services.json', 'r', encoding='utf-8') as file:
    json_data = json.load(file)
    for index, row in enumerate(json_data):
        service = next((item for item in services if item is not None and item.get("value", "") and item["value"] == row["category"][0]), None)
        if service is None:
            services.append({
                "img": row["imageSrc"],
                "name": ''.join([char.upper() if index == 0 else char.lower() for index, char in enumerate(row["category"][0])]),
                "value": row["category"][0],
                "packages": [row]
                })
        else:
            service["packages"].append(row)


with open('./data/output/services.json', 'w') as file:
    json.dump(services, file)
len(services)
```
